```python
import pandas as pd
from sqlalchemy import create_engine
import os
from dotenv import load_dotenv

# Загружаем переменные окружения из .env
load_dotenv()
```

```python
# Указываем данные для аутентификации (значения берутся из .env, в репозиторий не попадают)
db_config = {'user': os.environ['DB_USER'],       # имя пользователя
             'pwd': os.environ['DB_PASSWORD'],    # пароль
             'host': os.environ['DB_HOST'],
             'port': os.environ['DB_PORT'],       # порт подключения
             'db': os.environ['DB_NAME']          # название базы данных
             }

connection_string = 'postgresql://{}:{}@{}:{}/{}'.format(
    db_config['user'],
    db_config['pwd'],
    db_config['host'],
    db_config['port'],
    db_config['db'],
)
# Создаем движок
engine = create_engine(connection_string)
```

```python
query = '''
SELECT
    user_id,
    device_type_canonical,
    order_id,
    created_dt_msk AS order_dt,
    created_ts_msk AS order_ts,
    currency_code,
    revenue,
    tickets_count,
    EXTRACT(
        DAY FROM COALESCE(created_dt_msk - LAG(created_dt_msk) OVER(PARTITION BY user_id ORDER BY created_dt_msk), NULL)
    )::integer AS days_since_prev,
    p.event_id AS event_id,
    event_name_code AS event_name,
    event_type_main,
    service_name,
    region_name,
    city_name
FROM
    afisha.purchases AS p
JOIN
    afisha.events AS e ON p.event_id = e.event_id
JOIN
    afisha.city AS c ON e.city_id = c.city_id
JOIN
    afisha.regions AS r ON c.region_id = r.region_id
WHERE
    device_type_canonical IN ('mobile', 'desktop')
    AND event_type_main != 'фильм'
ORDER BY
    user_id
''' 
```

```python
df = pd.read_sql_query(query, con=engine)
```

```python
# Верхнеуровнево изучаем данные
def inspect_data(df):
    
    """
    DataFrame со статистикой по каждой колонке:
        
    - rows_cnt: количество непустых значений
    - isna_perc: процент пропущенных значений
    - cnt_nunique: количество уникальных значений
    - dtype: тип данных колонки
    - memory: занимаемая память в МБ
    - mode: самое частотное значение (первая мода)
            
            (только для числовых колонок)            
    - min: минимальное значение  
    - mean: среднее значение  
    - std: стандартное отклонение              
    - median: медианное значение  
    - max: максимальное значение    
    - sum: сумма значений  
    - sum_abs: сумма абсолютных значений, для возможной оценки между положительными и отрицательными значениями → sum_abs - sum 
    - neg_cnt: количество отрицательных значений в колонке
            
                
    Ограничения
    -----------
    Функция требует осторожности на больших данных (> 5 млн строк или > 500 колонок),
    особенно из-за операций nunique() и mode(), например:
    500 колонок при 1 млн строк (в сумме 500 млн ячеек)
    500 колонок + строковые/категориальные данные с высоким кардиналитетом (разнообразием значений)
            
    При недостатке памяти рекомендуется анализировать выборку колонок 
    >>> inspect_data(df.iloc[:, :20])  # только первые 20 колонок
    >>> # или
    >>> inspect_data(df[['col1', 'col2', 'col3']])
        
    """
    if df.empty:
        print("⚠️ DataFrame пуст")
        return pd.DataFrame()
    
    # Базовые метрики
    result = pd.DataFrame({
        'rows_cnt': df.count(),
        'isna_perc': (df.isna().mean() * 100).round(2),
        'cnt_nunique': df.nunique(),
        'dtype': df.dtypes.astype(str),
        'memory': (df.memory_usage(index=False) / 1024 / 1024).round(1),
        'mode': df.apply(lambda col: col.mode().iloc[0] if len(col.mode()) > 0 else pd.NA)        
    })
    # Разделяем статистики для числовых и нечисловых колонок
    numeric_cols = df.select_dtypes(include=['number', 'bool']).columns
    non_numeric_cols = df.select_dtypes(exclude=['number', 'bool']).columns

    result_numeric = result.loc[numeric_cols]
    result_non_numeric = result.loc[non_numeric_cols]
    
    if len(numeric_cols) > 0:
        stats = df[numeric_cols].agg(['min', 'mean', 'std', 'median', 'max', 'sum']).round(2).T
        # Добавляем сумму абсолютных значений и кол-во отриц. значений в колонке
        stats['sum_abs'] = df[numeric_cols].abs().sum().round(2)      
        # считает КОЛИЧЕСТВО, т.к. (df[numeric_cols] < 0) выдает True or False (1 или 0)
        stats['neg_cnt'] = (df[numeric_cols] < 0).sum()         
        # объединяем базовые метрики с числовыми статистиками
        result_numeric = result_numeric.join(stats)
    
    # сохраняем переменные
    return result_numeric, result_non_numeric

# Выводим информацию из датафрейма
# Распаковываем кортеж и выводим статистику
numeric_stats, non_numeric_stats = inspect_data(df)
display(non_numeric_stats)
display(numeric_stats)
```

```python
currency_rates = pd.read_csv('https://code.s3.yandex.net/datasets/final_tickets_tenge_df.csv')
# Проверяю что есть в датасете
currency_rates.head()
```

```python
currency_rates['data'] = pd.to_datetime(currency_rates['data'])

# Создаю предтаблицу
currency_pretable = df.merge(currency_rates, left_on='order_dt', right_on='data', how='left')

# Функция для конверсии валюты
def convertion(row):
    if row['currency_code'] == 'kzt':
        return row['revenue'] * row['curs'] / row['nominal']
    else:
        return row['revenue']
```

```python
# Применяю функцию к предтаблице
currency_pretable['revenue_rub'] = currency_pretable.apply(convertion, axis=1)
```

```python
# Применяю к исходному датафрейму
if len(df) == len(currency_pretable):
    df['revenue_rub'] = currency_pretable['revenue_rub']
else:
    raise ValueError('Количество строк отличается')
```

А вот интервал между заказами — сильный индикатор: у пользователей с 5+ заказами он почти вдвое короче (11.1 дня), чем у пользователей с 2-4 заказами (21.3 дня).

Корреляционный анализ (`phik`) подтвердил общую картину: сильнее всего с числом заказов связаны поведенческие метрики — частота заказов, среднее число билетов и средний чек (0.3-0.49), тогда как признаки первого заказа (устройство, регион, оператор, жанр, день недели) показывают крайне слабую связь (0.02-0.11). Отдельно отмечена взаимосвязь региона, оператора и жанра между собой (0.5-0.7) — вероятно, из-за территориальной специализации операторов и площадок.

Рекомендации:

Обстоятельства первого заказа (откуда пришёл пользователь, в какой день, через какого оператора) — слабый предиктор лояльности, поэтому не стоит переоценивать таргетинг по этим признакам. Гораздо информативнее ранние поведенческие сигналы, доступные уже после первых покупок: средний чек, число билетов в заказе и скорость повторного обращения.

Приоритетный сегмент для удержания — пользователи с первым заказом на 1 билет и низким средним чеком: это крупная группа (6146+ человек) с самой низкой долей повторных покупок (0.40). Для них стоит рассмотреть точечные стимулы после первой покупки — например, скидку или предложение на второй заказ в течение 1-2 недель, пока интервал между покупками ещё короткий у потенциально лояльных клиентов.

Сегмент с 2-3 билетами в заказе — самый благодарный по конверсии в повторные покупки (0.72); имеет смысл поощрять именно такое поведение (например, предлагать "билет для друга" при оформлении одиночной покупки).

Поскольку короткий интервал между заказами сильно коррелирует с лояльностью, стоит отслеживать пользователей, которые долго не возвращаются после первого заказа, и адресовать им реактивационные кампании (пуш-уведомления, персональные предложения) до того, как интервал станет слишком большим.

Отдельная рекомендация — обратить внимание на аномальные аккаунты с многократными заказами в сутки (исключённые 5% выборки): стоит проверить, не тратится ли на них маркетинговый бюджет, и рассмотреть внедрение анти-фрод фильтров на уровне продукта.
